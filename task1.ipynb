```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
train_df= pd.read_csv("train.csv")
test_df= pd.read_csv("test.csv")
```

```python
print("Columns in Train Data:", train_df.columns)
```

```python
print("Columns in Train Data:", test_df.columns)
```

```python
train_df=train_df[["GrLivArea","FullBath","BedroomAbvGr","SalePrice"]]
test_df=test_df[["GrLivArea","FullBath","BedroomAbvGr"]]
```

```python
train_df.head()
```

```python
X_train = train_df[["GrLivArea", "BedroomAbvGr", "FullBath"]]
X_test = test_df[["GrLivArea", "BedroomAbvGr", "FullBath"]]
```

```python
y_train = train_df["SalePrice"]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
predictions_df = pd.DataFrame({"PredictedSalePrice": y_pred})
```
